```python
import re
import pandas as pd
```

```python

f=open('WhatsApp Chat with Mummy.txt','r',encoding='utf-8')
data=f.read()
print(data)
```

```python
pattern='\d{1,2}\/\d{1,2}\/\d{2},\s\d{1,2}:\d{1,2}\s(?:AM|PM)\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
print(dates)
```

```python
df = pd.DataFrame({'user_message': messages, 'message_date': dates})
    # convert message_date type
    
#df['message_date'] = df['message_date'].str.strip()

# Replace the Unicode character with a space
df['message_date'] = df['message_date'].str.replace('\u202f', ' ')

# Convert to datetime
df['message_date'] = pd.to_datetime(df['message_date'], format='%m/%d/%y, %I:%M %p - ')

# Extract AM/PM and store in a new column
#df['am_pm'] = df['message_date'].dt.strftime('%p')

# Print the DataFrame
print(df)
df.rename(columns={'message_date': 'date'}, inplace=True)
```

```python
print(len(dates))
print(len(messages))
```

```python
users = []
messages = []
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
df
```

```python
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute

print(df)
```
